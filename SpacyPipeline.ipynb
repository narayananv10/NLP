```
import pandas as pd
import spacy
```

```
data = pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines=True)
df = pd.DataFrame(data)
```

```
sp = spacy.load('en_core_web_sm')
```

```
txt = df.reviewText[1]
```

```
sentence = sp(txt)
```

```
for word in sentence:
  print(word.pos_)
```

```
def SpacyPipeline(str):
  sentence = sp(str)
  nouns = []
  tokens = []
  lemma = []
  tags = []
  dep = []
  for i in sentence:
    tokens.append(i.text)
    lemma.append(i.lemma_)
    tags.append(i.pos_)
    dep.append(i.dep_)
  for n in sentence:
    if n.pos_ == 'NOUN':
      nouns.append(n.text)
  final = {'Tokens' : tokens,'Lemma' : lemma,'Tags' : tags, 'Dependency' : dep,'Nouns' : nouns}
  return final
```

```
SpacyPipeline(txt)
```
